Making the Scrawler

Importing the packages

```python
#importing the packages 
from selenium import webdriver
from selenium.webdriver.common.by import By
import random
import pandas as pd
import time
```

```python
# Initialize the web driver
driver = webdriver.Chrome()
```

```python
# List of links to scrape
links = [
    "https://www.myntra.com/women-sports-shoes",
    "https://www.myntra.com/women-sweaters-sweatshirts",
    "https://www.myntra.com/women-blazers-waistcoats",
    "https://www.myntra.com/women-jackets-coats",
    "https://www.myntra.com/ethnic-tops",
    "https://www.myntra.com/dresses?f=Gender%3Amen%20women%2Cwomen&sort=popularity"
]
```

```python
# Define the range for the number of items to scrape
MIN_ITEMS = 20
MAX_ITEMS = 30
target_items = random.randint(MIN_ITEMS, MAX_ITEMS)
scraped_items = []  # To store the scraped data
items_scraped_count = 0  # Counter to track the total scraped items
```

```python
# Function to extract product information from a list item element
def extract_product_info(li_element):
    try:
        # Get the product ID from the list item element
        product_id = li_element.get_attribute("id")
        
        # Locate the product meta information section
        product_meta_info = li_element.find_element(By.CLASS_NAME, "product-productMetaInfo")
        
        # Extract the brand name from the product meta information
        brand = product_meta_info.find_element(By.CLASS_NAME, "product-brand").text
        
        # Extract the product name from the product meta information
        product_name = product_meta_info.find_element(By.CLASS_NAME, "product-product").text
        
        # Extract the product price using a helper function
        price = _extract_product_price(product_meta_info)
        
        # Get the product image URL from the list item element
        image_url = li_element.find_element(By.TAG_NAME, "img").get_attribute("src")
        
        # Get the product link from the list item element
        product_link = li_element.find_element(By.TAG_NAME, "a").get_attribute("href")
        
        # Extract the rating and count of ratings using a helper function
        rating, count = _extract_rating_data(li_element)

        # Return a dictionary containing all extracted product information
        return {
            "id": product_id,
            "brand": brand,
            "product_name": product_name,
            "price": price,
            "rating": rating,
            "count": count,
            "image_url": image_url,
            "product_link": product_link,
        }
    except Exception as e:
        # Print an error message if an exception occurs during extraction
        print(f"Error extracting product info: {e}")
        return None  # Return None if extraction fails
```

```python
# Helper function to extract product price information
def _extract_product_price(product_meta_info):
    # Initialize a dictionary to hold discounted and original prices
    price = {"Discounted": 0, "Original": 0}
    try:
        # Attempt to extract the discounted price
        price["Discounted"] = product_meta_info.find_element(By.CLASS_NAME, "product-discountedPrice").text.replace("Rs. ", "").strip()
        
        # Attempt to extract the original price
        price["Original"] = product_meta_info.find_element(By.CLASS_NAME, "product-strike").text.replace("Rs. ", "").strip()
    except:
        # If the original price is not found, try to extract it from a different element
        price["Original"] = product_meta_info.find_element(By.TAG_NAME, "span").text.replace("Rs. ", "").strip()
    return price  # Return the price dictionary
```

```python
# Helper function to extract rating data from a list item element
def _extract_rating_data(li_element):
    try:
        # Locate the ratings container within the list item element
        ratings_container = li_element.find_element(By.CLASS_NAME, "product-ratingsContainer")
        
        # Extract the rating value from the ratings container
        rating_value = ratings_container.find_element(By.TAG_NAME, "span").text
        
        # Extract the count of ratings, splitting the text to get the relevant part
        ratings_count = ratings_container.find_element(By.CLASS_NAME, "product-ratingsCount").text.split('|')[-1].strip()
        
        return rating_value, ratings_count  # Return the rating value and count
    except:
        return None, None  # Return None if extraction fails
```

```python
# Shuffle the links to randomize the scraping order
random.shuffle(links)

# Loop through each link
for link in links:
    if items_scraped_count >= target_items:
        break  # Stop scraping if we've reached the target items

    driver.get(link)
    time.sleep(1)  # Wait for the page to load

    # Find all product <li> elements
    product_elements = driver.find_elements(By.CLASS_NAME, "product-base")
    
    # Shuffle the products to randomize the scraping order
    random.shuffle(product_elements)

    for product in product_elements:
        if items_scraped_count >= target_items:
            break  # Stop scraping if we've reached the target items
        
        product_info = extract_product_info(product)
        if product_info:
            scraped_items.append(product_info)
            items_scraped_count += 1

```

```python
# Function to extract comments and associated ratings
def extract_comments(p_id, product_review_url):
    """
    Extracts reviews and associated ratings for a given product ID.
    Parameters:
        p_id (str): Product ID
        product_review_url (str): URL to the product's review page
    """
    try:
        # Set an implicit wait for page elements
        driver.implicitly_wait(5)
        
        # Locate review elements
        rating_elements = driver.find_elements(By.CLASS_NAME, "user-review-starRating")
        review_container = driver.find_element(By.CLASS_NAME, "detailed-reviews-userReviewsContainer")
        reviews = review_container.find_elements(By.CLASS_NAME, "user-review-reviewTextWrapper")
        date_elements = driver.find_elements(By.CLASS_NAME, "user-review-left")

        # Extract a maximum of 5 reviews
        for i in range(min(5, len(reviews))):
            rating = rating_elements[i].text.strip() if i < len(rating_elements) else "N/A"
            comment = reviews[i].text.strip()
            date = date_elements[i].find_elements(By.TAG_NAME, "span")[1].text.strip()

            # Append review data to the review_data list
            review_data.append({
                'product_id': p_id,
                'cust_rating': rating,
                'comment': comment,
                'date': date
            })

    except Exception as e:
        print(f"Error scraping reviews for product ID {p_id}: {e}")

```

```python
# Function to extract average ratings
driver = webdriver.Chrome()
def extract_ratings(p_id, product_review_url):
    """
    Extracts the average rating for a given product ID.
    Parameters:
        p_id (str): Product ID
        product_review_url (str): URL to the product's review page
    """
    try:
        # Set a longer implicit wait for elements
        driver.implicitly_wait(5)
        
        # Locate the average rating container and extract the value
        avg_container = driver.find_element(By.CLASS_NAME, "index-averageRating")
        avg_ratings = avg_container.find_element(By.TAG_NAME, "span").text
        
        # Append average rating data to the ratings_data list
        ratings_data.append({
            "product_id": p_id,
            "avg": avg_ratings
        })

    except Exception as e:
        print(f"Error scraping ratings for product ID {p_id}: {e}")
```

```python
# Convert the scraped data to a pandas DataFrame and save to a CSV file
df = pd.DataFrame(scraped_items)
df.to_csv("scraped_data.csv", index=False)

print(f"Scraped {items_scraped_count} items (randomly selected). Data saved to 'scraped_data.csv'.")
```

```python
df = pd.read_csv('scraped_data.csv')

ratings_data = []
review_data = []
review_url = 'https://www.myntra.com/reviews/'

#Iterates through all the product_id in the stored csv file and creates a link form each product.
p_ids = df['id']
for p_id in p_ids:
    product_review_url = review_url + str(p_id)
    driver.get(product_review_url)
    extract_comments(p_id, product_review_url)
    extract_ratings(p_id, product_review_url)

df_reviews = pd.DataFrame(review_data)
df_reviews.to_csv('reviews.csv', index=False)
print("reviews.csv saved successfully.")

df_ratings = pd.DataFrame(ratings_data)
df_ratings.to_csv('ratings.csv', index=False)
print("ratings.csv saved successfully")
```
